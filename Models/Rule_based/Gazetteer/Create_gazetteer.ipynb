```python
import requests
from bs4 import BeautifulSoup

url = "https://www.britannica.com/topic/list-of-philosophers-2027173"
html = requests.get(url).content
data = BeautifulSoup(html, "html.parser")

philosophers = []
for ultag in data.find_all('ul', {'class': 'topic-list'}):
    for litag in ultag.find_all('li'):
        #print(litag.text)
        philosophers.append(litag.text)
        
print(philosophers)
```

```python
import json
print(json.dumps(philosophers))
```
